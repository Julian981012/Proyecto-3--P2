Esta parte del código nos sirve para poder subir los datos de control y parkinson que se encuentran subidos en el Drive y poder trabajar con ellos posteriormentre.

```python
def calcular_energia(senal):
   potencia = np.sum(senal ** 2)
   energia = np.mean(potencia)
   return energia
```

Primero, calculamos la potencia sumando el cuadrado de cada punto de la señal para después pasar a calcular la energía como la media de esta potencia.

```python
def energia_epocas(senal):
   energias_epocas = []

   for columna in range(senal.shape[1]):
       epoca = senal[:, columna]
       energia_epoca = calcular_energia(epoca)
       energias_epocas.append(energia_epoca)

   energia_prom = np.mean(energias_epocas)

   return energia_prom
```

Definimos la función que calcula la energía promedio de las épocas de una señal. Para cada época en la señal, se calcula la energía y se almacena en una lista y luego calculamos la energía promedio de todas las épocas.

```python
def calcular_energia_epocas(senal):
    energias = []

    for columna in range(senal.shape[1]):
        epoca = senal[:, columna]
        energia_epoca = calcular_energia_signal(epoca)
        energias.append(energia_epoca)

    energia_promedio = np.mean(energias)

    return energia_promedio

def calcular_energia_signal(senal):

    potencia_instantanea = np.sum(senal ** 2)  # Calcula la potencia instantánea
    energia = np.mean(potencia_instantanea)  # Calcula la energía como la media de la potencia

    return energia


rutas_control = glob.glob('/content/drive/MyDrive/control/control/*.mat')
datos_control = pd.DataFrame(columns=['Canal', 'Sujeto', 'Energia'])

for i, archivo in enumerate(rutas_control):
    datos_mat_control = sio.loadmat(archivo)
    senales_control = datos_mat_control['data']
    for canal in range(8):
        energia_canal = calcular_energia_epocas(senales_control[canal, :, :])
        datos_control = pd.concat([datos_control, pd.DataFrame({'Canal': f'Canal {canal+1}', 'Sujeto': i+1, 'Energia': [energia_canal]})], ignore_index=True)


datos_control['Sujeto'] = datos_control['Sujeto'].astype(int)

tabla_datos_control = pd.DataFrame.pivot(datos_control, index='Sujeto', columns='Canal', values='Energia')
tabla_datos_control.columns.name = None

rutas_parkinson = glob.glob('/content/drive/MyDrive/parkinson/parkinson/*.mat')
datos_parkinson = pd.DataFrame(columns=['Canal', 'Sujeto', 'Energia'])

for i, archivo in enumerate(rutas_parkinson):
    datos_mat_parkinson = sio.loadmat(archivo)
    senales_parkinson = datos_mat_parkinson['data']
    for canal in range(8):
        energia_canal = calcular_energia_epocas(senales_parkinson[canal, :, :])
        datos_parkinson = pd.concat([datos_parkinson, pd.DataFrame({'Canal': f'Canal {canal+1}', 'Sujeto': [i+1], 'Energia': [energia_canal]})], ignore_index=True)

# Convertir la columna 'Sujeto' a tipo numérico
datos_parkinson['Sujeto'] = datos_parkinson['Sujeto'].astype(int)

tabla_datos_parkinson = pd.DataFrame.pivot(datos_parkinson, index='Sujeto', columns='Canal', values='Energia')
tabla_datos_parkinson.columns.name = None
```

Aquí, pasamos a calcular la energía promedio de las señales EEG utilizando la funcion anterior para sujetos de un grupo control y un grupo de pacientes con Parkinson. Para cada sujeto y para cada canal de la señal EEG, se calcula la energía promedio de las épocas de la señal. Después, estos valores se organizan en tablas para el grupo control y el grupo de pacientes con Parkinson, lo que permite comparar la energía promedio entre los grupos y los canales de la señal EEG.

Imprimimos los diferentes DataFrames para poder ver y revisar si estábamos haciendo bien los procedimientos y cargando los datos, ya que durante la realización del código varias veces nos salia un DataFrame en blanco, sin datos.

```python
def prueba_normalidad(data, nombre):
    print(nombre)
    for canal in data.columns:
        stat, p = stats.shapiro(data[canal])
        print(f'{canal}: Estadístico={stat:.3f}, p={p:.3f}')

print('Datos Control')
prueba_normalidad(tabla_datos_control, 'Control:')

print('\nDatos Parkinson')
prueba_normalidad(tabla_datos_parkinson, 'Parkinson:')
```

Los resultados indican que, debido a la falta de normalidad en los datos, la aplicación directa de la prueba t podría no ser adecuada, ya que esta prueba asume que los datos siguen una distribución normal. Dado que la hipótesis nula en la prueba t supone que no hay diferencia significativa entre las medias de dos grupos, la falta de normalidad podría afectar la validez de esta prueba. Por lo tanto, sería prudente considerar el uso de pruebas no paramétricas, que no dependen de la suposición de normalidad, o explorar la posibilidad de transformar los datos para cumplir con los supuestos de la prueba t, si se considera necesario su uso.

```python
def prueba_homocedasticidad(datos_control, datos_parkinson):
   for canal in datos_control.columns:
       stat, p = stats.levene(datos_control[canal], datos_parkinson[canal])
       print(f'{canal}: Estadístico={stat:.3f}, p={p:.3f}')

prueba_homocedasticidad(tabla_datos_control, tabla_datos_parkinson)
```

Los resultados de las pruebas de homogeneidad, en particular la prueba de Levene para este caso, sugieren que las variaciones entre los diferentes canales son similares, ya que los valores de p altos indican que las diferencias no son significativas. Esta similitud en las variaciones es importante, ya que nos permite comparar los canales de manera más confiable, sin preocuparnos de que las diferencias observadas se deban a variaciones muy diferentes entre los canales en lugar de las condiciones reales que estamos tratando de medir.

```python
def prueba_mann_whitney(datos_control, datos_parkinson):
   for canal in datos_control.columns:
       stat, p = stats.mannwhitneyu(datos_control[canal], datos_parkinson[canal])
       print(f'{canal}: Estadístico={stat:.3f}, p={p:.3f}')

prueba_mann_whitney(tabla_datos_control, tabla_datos_parkinson)
```

Los resultados de la prueba de Mann-Whitney para cada canal muestran valores de p altos, lo que sugiere que no hay evidencia suficiente para rechazar la hipótesis nula de que las distribuciones de los dos grupos son iguales. Estos resultados son coherentes con los hallazgos anteriores de homogeneidad de varianzas, lo que respalda la idea de que las diferencias observadas entre los canales podrían no deberse a variaciones muy diferentes entre ellos. En resumen, los resultados de la prueba de Mann-Whitney indican que no hay diferencias significativas entre los canales en términos de las mediciones realizadas.

# **Analisis general:**

Los análisis estadísticos realizados tuvieron como objetivo validar la comparación de la energía de las señales EEG y evaluar la confiabilidad de los resultados. Inicialmente, la prueba de normalidad (Shapiro-Wilks) indicó que los datos no seguían una distribución normal, lo que es necesario para pruebas paramétricas como la prueba t. Posteriormente, la prueba de Pearson confirmó la independencia entre los canales homólogos de ambos grupos. La prueba de Levene evaluó la igualdad de varianzas entre grupos, un requisito fundamental para ciertos análisis estadísticos. Dado que no se cumplían todos los requisitos para la prueba t, se utilizó la prueba U de Mann-Whitney, que no requiere distribución normal de los datos. Los valores de p altos obtenidos en todas las pruebas indicaron que no había diferencias estadísticamente significativas entre los grupos para cada canal. En resumen, los resultados sugieren que los canales evaluados no son efectivos para discriminar entre los grupos estudiados en términos de la energía promedio, posiblemente debido a la escasa variabilidad entre los sujetos en la muestra o a que la energía promedio de los canales no es una medida sensible para detectar las diferencias entre los grupos.

# **Conclusiones:**



*   Estos hallazgos sugieren que la medida utilizada para la energía de los canales puede no ser sensible para detectar diferencias entre los grupos, posiblemente debido a la limitada variabilidad entre los participantes en el estudio. En resumen, estos resultados enfatizan la importancia de considerar cuidadosamente las medidas utilizadas en la investigación y la necesidad de explorar otras variables que puedan ofrecer una mejor discriminación entre los grupos estudiados.

*   La falta de evidencia de una distribución normal en los datos junto con la homogeneidad de varianzas evaluada, respaldan la decisión de utilizar pruebas no paramétricas. La aplicación de la prueba U de Mann-Whitney indicó que no había diferencias estadísticamente significativas entre los grupos para cada canal, lo que sugiere que los canales evaluados no son efectivos para discriminar entre los grupos estudiados en términos de la energía promedio.
